```python
import os
import nltk
data = "/content/Data/Data"
os.chdir(data)
os.listdir(data)
```

a. Preprocessing
1. Converting to lower case

```python
lower_case = os.path.join(data,'lc_files')
os.makedirs(lower_case,exist_ok=True)
data_folder = [f for f in os.listdir(data) if f.endswith(".txt")]
for files in data_folder:
  file = os.path.join(data,files)
  with open(file,'r',encoding="ISO-8859-1")as f:
    text = f.read()
  convert_to_lower_case = text.lower()
  lower_text_file = os.path.join(lower_case,files)
  with open(lower_text_file,'w',encoding="ISO-8859-1")as f:
    f.write(convert_to_lower_case)
  print(f"converted output file :{lower_text_file} ")
```

2. Stopwords

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('punkt')
data = "/content/Data/Data/lc_files"
data_dir = '/content/Data/Data'
stop_words = os.path.join(data_dir,'stopword_files')
os.makedirs(stop_words,exist_ok=True)
stopwords = set(stopwords.words('english'))
data_folder = [f for f in os.listdir(data) if f.endswith(".txt")]
for files in data_folder:
  file = os.path.join(data,files)
  with open(file,'r',encoding="ISO-8859-1")as f:
    read_lowercase_file = f.read()
  words = nltk.word_tokenize(read_lowercase_file)
  filtered_words = [word for word in words if word.lower() not in stopwords]
  sw_file = os.path.join(stop_words,files)
  with open(sw_file,'w',encoding="ISO-8859-1")as f:
    f.write(' '.join(filtered_words))
  print(f"converted output file :{sw_file} ")
```

3. Punctuation

```python
import string
data = "/content/Data/Data/stopword_files"
data_dir = '/content/Data/Data'
punctuation = os.path.join(data_dir,'punctuation_files')
os.makedirs(punctuation,exist_ok=True)
data_folder = [f for f in os.listdir(data) if f.endswith(".txt")]
translator = str.maketrans('','',string.punctuation)
for files in data_folder:
  file = os.path.join(data,files)
  with open(file,'r',encoding="ISO-8859-1")as f:
    read_sw_file = f.read()
  remove_punctuation = read_sw_file.translate(translator)
  punk_file = os.path.join(punctuation,files)
  with open(punk_file,'w',encoding="ISO-8859-1")as f:
    f.write(remove_punctuation)
  print(f"converted output file :{punk_file} ")
```

4. Single Charecters

```python
data = "/content/Data/Data/punctuation_files"
data_dir = '/content/Data/Data'
single_charecter = os.path.join(data_dir,'single_charecter_files')
os.makedirs(single_charecter,exist_ok=True)
data_folder = [f for f in os.listdir(data) if f.endswith(".txt")]
for files in data_folder:
  file = os.path.join(data,files)
  with open(file,'r',encoding="ISO-8859-1")as f:
    read_punk_file = f.read()
    words = read_punk_file.split()
  filtered_words = [word for word in words if len(word)>1]
  sc_file = os.path.join(single_charecter,files)
  with open(sc_file,'w',encoding="ISO-8859-1")as f:
    f.write(' '.join(filtered_words))
  print(f"converted output file :{sc_file} ")
```

5. Lemmatization

```python
data = '/content/Data/Data/single_charecter_files'
data_dir = '/content/Data/Data'
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
lemmatizer = WordNetLemmatizer()
lemmatization = os.path.join(data_dir,'lemmatization_files')
os.makedirs(lemmatization,exist_ok=True)
def wordnet_pos(tag):
  if tag.startswith('J'):
    return wordnet.ADJ
  elif tag.startswith('V'):
    return wordnet.VERB
  elif tag.startswith('N'):
    return wordnet.NOUN
  elif tag.startswith('R'):
    return wordnet.ADV
  else:
    return wordnet.NOUN
data_folder = [f for f in os.listdir(data) if f.endswith(".txt")]
for files in data_folder:
  file = os.path.join(data,files)
  with open(file,'r',encoding="ISO-8859-1")as f:
    read_sc_file = f.read()
    lm_words = nltk.word_tokenize(read_sc_file)
    tags = nltk.pos_tag(lm_words)
    lemmatized_words = [lemmatizer.lemmatize(word,pos=wordnet_pos(tag)) for word,tag in tags]
    lm_file = os.path.join(lemmatization,files)
    with open(lm_file,'w',encoding="ISO-8859-1")as f:
      f.write(' '.join(lemmatized_words))
    print(f"converted output file :{lm_file} ")
```

6. Converting numbers

```python
import re
import inflect
data = '/content/Data/Data/lemmatization_files'
data_dir = '/content/Data/Data'
numbers = os.path.join(data_dir,'numbers_to_text_files')
os.makedirs(numbers,exist_ok=True)
a = inflect.engine()
data_folder = [f for f in os.listdir(data) if f.endswith(".txt")]
for files in data_folder:
  file = os.path.join(data,files)
  with open(file,'r',encoding="ISO-8859-1")as f:
    read_lm_file = f.read()
    text = read_lm_file.split()
    number_to_text = []
    for word in text:
      n_c = re.sub(r'\D','',word)
      if n_c.isdigit():
        numb =int(n_c)
        if 0<=numb<-10**6:
          number_to_text.append(a.number_to_words(numb))
        else:
          number_to_text.append(word)
      else:
        number_to_text.append(word)
    nct_file = os.path.join(numbers,files)
    with open(nct_file,'w',encoding="ISO-8859-1")as f:
      f.write(' '.join(number_to_text))
    print(f"converted output file :{nct_file} ")
```

b. Extracting TF-IDF

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
data = "/content/Data/Data/numbers_to_text_files"
data_dir = '/content/Data/Data'
data_folder = [f for f in os.listdir(data) if f.endswith(".txt")]
docs =[]
file_name=[]
for files in data_folder:
  filepath = os.path.join(data,files)
  with open(filepath,'r',encoding="ISO-8859-1")as f:
    docs.append(f.read())
    file_name.append(files)
  k = 20
  vect = TfidfVectorizer(max_features = k)
  tfidf = vect.fit_transform(docs)
  feature_name = vect.get_feature_names_out()
  data_tfidf = pd.DataFrame(tfidf.toarray(),columns=feature_name,index=file_name)
  tfidf_file = os.path.join(data_dir,'tfidf.csv')
  data_tfidf.to_csv(tfidf_file,encoding='ISO8859-1')
  print(data_tfidf.head())
```

d. Top 10 retrival results

```python
import numpy as np
data = "/content/Data/Data/numbers_to_text_files"
data_folder = [f for f in os.listdir(data) if f.endswith(".txt")]
docs=[]
for files in data_folder:
  filepath = os.path.join(data,files)
  with open(filepath,'r',encoding="ISO-8859-1")as f:
    docs.append(f.read())
queries = ["Once upon a time there were three little pigs who left thier mummy and daddy to see the world.",
           "There once lived a poor tailor who had a son called Aladdin, a careless, idle boy who would do nothing but play all day long in the streets with little boys like himself."]
vect = TfidfVectorizer()
tfidf = vect.fit_transform(docs)
query_matrix = vect.transform(queries)
similarity = cosine_similarity(query_matrix,tfidf)
for i, query in enumerate(queries):
  top_relevenat_file = np.argsort(similarity[i])[::-1][:10]
  print(f"Top 10 documents for query: \"{query}\"")
  for index in top_relevenat_file:
    print(data_folder[index])
```
